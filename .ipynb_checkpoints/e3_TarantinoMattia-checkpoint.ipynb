```python
import numpy as np
import pandas as pd
```

```python
p_0 = 0.02
```

```python
r_s = 30
R = r_s * 10

Intervallo = R 

h_0 = (R - 0)/3

x = []

for i in range(0,4):

    y = i*h_0
    x.append(y)

print(x)
```

```python
def f(r, p_0 = 0.02, r_s = 30):

    d = r/r_s
    f = 4*np.pi*(r * p_0 * r_s)/(1 + d)**2
    return f
```

```python
def integrale_M_R(r, h):

    i = 0
    
    I =(3/8)*h* (f(r[i]) + 3 * f(r[i + 1]) + 3 * f(r[i + 2]) + f(r[i + 3]))
    return I
def integrale_M_R_Senza_Pi(r):

    i = 0
    
    I =(f(r[i]) + 3 * f(r[i + 1]) + 3 * f(r[i + 2]) + f(r[i + 3]))
    return I
```

```python
def intervallo_x(N, h):

    x = []

    for i in range(N + 1):

        y = i*h
        x.append(y)
    return x
```

```python
I_0 = integrale_M_R(x , h_0)
print(I_0)
```

```python
N = 3
k = 0
j = 0
I = 0
I_2 = 1
Err_M = 1
while(Err_M > 1e-5):

    I = 0
    I_1 = I_2

    k = k + 1
    
    h = R/N
    j = int(h_0/h)
    x = intervallo_x(N,h)
    #print(x)
    #print(f"Ciclo {j}:")
    #print(f"h: {h}\n\nN:{N}\n\nNh:{N*h}")
    for i in range(j):
        
        x_1 = x[3*i : 3*i + 4: 1]
        I += integrale_M_R_Senza_Pi(x_1)
        #print(f"Intervallo {i + 1}: {len(x_1)}\n")        

    I = (3/8)*h* I
    #print(I)
    I_2 = I
    N = 2*N

    Err_M = np.abs(I_2 - I_1)/np.abs(I_2)
print(f"L'integrale viene: {I_2} +/- {Err_M * 100:.5f}%")
print(f"Tale Risultato è stato ottenuto con un passo h: {h}")
I_p_2 = I_2
Err_M_1 = Err_M
```
