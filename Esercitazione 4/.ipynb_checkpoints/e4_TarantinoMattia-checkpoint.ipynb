```python
import numpy as np 
```

```python
def df(x, y):

    return x**3 - 1000*y
```

```python
def stima_wi(h = 0.002, r = 1):

    y = ((r*h)**4)/4 - 50*(r*h)**5
    return y

epsilon = 1e-9
h = 0.002
def predictor(x, w, h = 0.002):

    w5 = 55*df(x[-1], w[-1])
    w5 -= 59*df(x[-2], w[-2])
    w5 += 37*df(x[-3], w[-3])
    w5 -= 9*df(x[-4], w[-4])
    w5 = w[-1] + (h/24)*(w5)
    return w5

def corrector(x,w, h = 0.002):
    
    w5new = w[- 1]
    err = [1]
    while(err[-1] > epsilon):

        w5new = ( 9 * df(x[- 1], w[-1])) + (19 * df(x[-2], w[-2])) + (-5 * df(x[-3], w[-3])) + df(x[-4] , w[-4])
        w5new = w[-2] + (h/24)*w5new

        err = np.append(np.abs(w5new - w[-1]), err)
        
        w[-1] = w5new
    #print(f"Errore: {err}")
    return w5new, err[0]
```

```python
w = []

for i in range(0,4):

    w.append(stima_wi(r = i))

wi = np.array(w)

print(wi)

x = []
for i in range(4):
    x.append(i * h)

xs = np.array(x)

w_5 = predictor(xs, wi)
print(w_5)

err = []

while (xs[-1] < 0.8 + h):

    wi = np.append(wi, predictor(xs, wi))
    #print(f"\n\nPredictor: {wi[-1]}\n\n")
    wi[-1], e = corrector(xs, wi)
    err.append(e)
    xs = np.append(xs, xs[-1] + h)
    #print(f"Corrector: {wi[-1]}\n\n")
    print(f"X: {xs[-1]}")
err = np.array(err)
print(err)
```
