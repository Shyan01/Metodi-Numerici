```python
import numpy as np
import pandas as pd
import sys

epsilon = 1e-3
```

```python

file_name = "Esercitazione 5.xlsx"
df = pd.read_excel(file_name)

M = df.to_numpy()
S = M[:, -1]
M = np.delete(M, -1, axis = -1)

col, rows = np.shape(M)
w = 1
```

```python
def sol(A,x):

    x_col = x.reshape(-1, 1)

    sol = np.dot(A, x_col)
    sol = sol.flatten()
    
    return sol

def Gauss_Seidel(A, B, X):
    X_0 = np.zeros(len(X))
    err = np.abs(X_0[0] - X[0])
    count = 0
    while(err > epsilon):
        X_0 = np.copy(X)

        for i in range( rows):
            Sum = 0
            
            for j in range(i):

                Sum += A[i][j] * X_0[j]
                
            for k in range(i + 1, col):

                Sum += A[i][k] * X[k]

            X[i] = (B[i] - Sum)*(1/A[i][i]) 
            err = np.max(np.abs(X[i] - X_0[i]))
        count += 1
    return X, count

def Gauss_Seidel_w(A, B, X, w):
    X_0 = np.zeros(len(X))
    err = float('inf')
    count = 0
    while(err > epsilon):
        X_0 = np.copy(X)
        count += 1
        for i in range( rows):
            
            err_1 = []
            err_1 = np.array(err_1)
            Sum = 0
            
            for j in range(i):

                Sum += A[i][j] * X[j]
                
            for k in range(i + 1, col):

                Sum += A[i][k] * X_0[k]

            X[i] = (1-w) * X_0[i] + (w/A[i][i])*(B[i] - Sum)

            err_1 = np.append(err_1, np.abs(X[i] - X_0[i]))
            err = np.max(err_1)

    return X, count
```

```python

Verifica = []
Somme = []
for i in range(len(M)):

    v = M[i]
    s = np.sum(np.abs(v)) - np.abs(M[i][i])
    Somme.append(s)
    if(np.abs(M[i][i]) > s):
        Verifica.append(True)
    else:
        Verifica.append(False)

mask = False
check = Verifica[mask]
print(check)
Somme = np.array(Somme)
print(Somme)
if not check:
    print("Matrice non rispetta la condizione sulla diagonale")
    sys.stop()
x_i = np.ones(len(M[1]))
print(x_i)

X ,count = Gauss_Seidel(M, S, x_i)
X = np.array(X)
Soluzione = sol(M,X)
for i in range(len(X)):
    
    print(f"X: {X[i]:.3f}")
print(f"Soluzione: {Soluzione}\n\nIterazioni: {count}")
```
