```python
import pandas as pd
import numpy as np
```

```python
def Delta(y):

    L = len(y) - 1
    for i in range(L):

        y[i] = y[i+1] - y[i]
    d = y[:-1]
    return d
```

```python
def coefficent( r, i, c = 1):

    c = c*(r-i+1)/i
    return c
```

```python
def f_x(x_j, h, y_j, x):

    a = 1/h
    b = x_j * a
    delta = Delta(y_j)
    d_1 = delta[0]
    delta_2 = Delta(delta)
    d_2 = delta_2[0]
    delta_3 = Delta(delta_2)
    d_3 = delta_3[0]
    c = d_3/d_2
    d = (1 + b - 1/c)
    f = d_1/(a * c * d * d_2) + ((b**2)/2*a*d) - ((2/3)* 1/a*d) + ((a/2*d)*x) - x
    return f
```

```python
def df_x(x_j, h, y_j, x):

    a = 1/h
    b = x_j * a
    delta = Delta(y_j)
    d_1 = delta[0]
    delta_2 = Delta(delta)
    d_2 = delta_2[0]
    delta_3 = Delta(delta_2)
    d_3 = delta_3[0]
    c = d_3/d_2
    d = (1 + b - 1/c)
    df = (a/d) * x - 1
    return df
```

```python
def Newton_Rapson(x_j, h, y_j, x, i = 0):

    epsilon = 1e-6
    k = i
    k += i
    f = f_x(x_j, h, y_j, x)
    df = df_x(x_j, h, y_j, x)
    x_1 = x - f/df
    print(x_1)

    if abs(x_1 - x) < epsilon:
        
        return x_1 , i
    else:
        Newton_Rapson(x_j, h, y_j, x_1, k)
```
