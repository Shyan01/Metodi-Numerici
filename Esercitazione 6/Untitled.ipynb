```python
import numpy as np
import random as rand
import matplotlib.pyplot as plt
import time as tm
```

```python
h = 1/4
def f(x):

    return np.sqrt(x)


def P(x,A):

    p = 1/A
    p *= f(x)
    return p


def Integrale(A ,B, N, f): 

    h = B - A
    h = h/N

    I = f(A) + f(B)

    I += 4 * f(A+h)
    
    for i in range(1,int(N/2)):
        x = A +2*i*h
        I += 2 * f(x)

        x += h
        I += 4 * f(x)
        
    I *= h/3
    return I 

def Media_F(x , A):

    M = x * f(x)/A
    return M


def distribuzione_prob_analitica(N):

    X = np.zeros(N)
    rand.seed(int(tm.time()))
    for i in range(N):

        r = rand.random()
        X[i] = np.cbrt(r**2)

    return X
```

```python
n = 100000
A = Integrale(0, 1 , n, f)

print(A)
```
